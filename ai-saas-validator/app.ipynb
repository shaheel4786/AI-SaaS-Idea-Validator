```python
# Create agents.py
agents_code = """
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def run_agent(role_prompt, idea):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": role_prompt},
            {"role": "user", "content": idea}
        ],
        temperature=0.7
    )
    return response.choices[0].message.content


def market_agent(idea):
    prompt = '''
    You are a Market Research Analyst.
    Provide:
    - Industry trends
    - TAM, SAM, SOM estimates
    - Target audience
    - Growth rate
    '''
    return run_agent(prompt, idea)


def competitor_agent(idea):
    prompt = '''
    You are a Competitor Research Analyst.
    Identify:
    - Direct competitors
    - Indirect competitors
    - Features
    - Pricing models
    '''
    return run_agent(prompt, idea)


def monetization_agent(idea):
    prompt = '''
    You are a SaaS Monetization Strategist.
    Suggest:
    - Revenue model
    - Pricing tiers
    - Upsell strategy
    - Revenue assumptions
    '''
    return run_agent(prompt, idea)


def risk_agent(idea):
    prompt = '''
    You are a Risk Analysis Expert.
    Identify:
    - Technical risks
    - Market risks
    - Regulatory risks
    Provide overall risk score (Low/Medium/High).
    '''
    return run_agent(prompt, idea)
"""

with open("agents.py", "w") as f:
    f.write(agents_code)


# Create orchestrator.py
orchestrator_code = """
from agents import market_agent, competitor_agent, monetization_agent, risk_agent

def run_full_analysis(idea):
    market = market_agent(idea)
    competitors = competitor_agent(idea)
    monetization = monetization_agent(idea)
    risks = risk_agent(idea)

    return {
        "market": market,
        "competitors": competitors,
        "monetization": monetization,
        "risks": risks
    }
"""

with open("orchestrator.py", "w") as f:
    f.write(orchestrator_code)


# Create report_generator.py
report_code = """
def generate_final_report(results):
    report = f'''
# 🚀 AI SaaS Idea Validation Report

## 📈 Market Research
{results['market']}

## 🏢 Competitor Landscape
{results['competitors']}

## 💰 Monetization Strategy
{results['monetization']}

## ⚠ Risk Assessment
{results['risks']}

## ✅ Final Recommendation
Further validation and MVP testing is recommended before launch.
'''
    return report
"""

with open("report_generator.py", "w") as f:
    f.write(report_code)


# Create app.py
app_code = """
import streamlit as st
from orchestrator import run_full_analysis
from report_generator import generate_final_report

st.set_page_config(page_title="AI SaaS Idea Validator", layout="wide")

st.title("🚀 AI SaaS Idea Validator")
st.markdown("Validate your startup idea using Multi-Agent AI System")

idea = st.text_area("Enter Your Startup Idea")

if st.button("Validate Idea"):
    if idea:
        with st.spinner("Running Multi-Agent Analysis..."):
            results = run_full_analysis(idea)
            report = generate_final_report(results)

        st.success("Analysis Complete!")

        tabs = st.tabs(["Market", "Competitors", "Monetization", "Risk", "Final Report"])

        with tabs[0]:
            st.write(results["market"])

        with tabs[1]:
            st.write(results["competitors"])

        with tabs[2]:
            st.write(results["monetization"])

        with tabs[3]:
            st.write(results["risks"])

        with tabs[4]:
            st.markdown(report)
    else:
        st.warning("Please enter a startup idea.")
"""

with open("app.py", "w") as f:
    f.write(app_code)

print("All project files created successfully!")
```
